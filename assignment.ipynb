```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline
```

```python
x = np.random.randint(1000, size=50)
y = np.random.randint(1000, size=50)
```

```python
fig, ax = plt.subplots()
#subplots(2,2,1)
c = (x+y)%2 == 0
ax.scatter(x[c], y[c], c="blue")
ax.scatter(x[~c], y[~c], c="red")

plt.show()
```

```python
x1=np.linspace(0,5,100)
```

```python
y1 = 3*x1 + 0.5
```

```python
 y2 = 5*np.sqrt(x1)
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()

ax.plot(x1, y1,'k--*',label='Inline label')
#ax.legend()
#ax.legend(loc='upper left',)
ax.plot(x1, y2, c="yellow",label='Inline label')
ax.set_xscale("log")
ax.legend(loc='upper left',)

plt.show()
```
