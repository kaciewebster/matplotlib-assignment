```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline
```

```python
x = np.random.randint(1000, size=50)
y = np.random.randint(1000, size=50)
```

```python
fig, ax = plt.subplots()
c = (x+y)%2 == 0
ax.scatter(x[c], y[c], c="blue")
ax.scatter(x[~c], y[~c], c="red")

plt.show()
```
